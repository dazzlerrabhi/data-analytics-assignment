# Data Cleaning

```python
import pandas as pd
```

### Load Datasets

```python
user_details = pd.read_excel(r'Assignment.xlsx', sheet_name = 'UserDetails.csv')

cooking_sessions = pd.read_excel(r'Assignment.xlsx', sheet_name = 'CookingSessions.csv')

order_details = pd.read_excel(r'Assignment.xlsx', sheet_name = 'OrderDetails.csv')
```

## Step 1: Data Cleaning

```python
# Remove duplicates

user_details = user_details.drop_duplicates()

cooking_sessions = cooking_sessions.drop_duplicates()

order_details = order_details.drop_duplicates()
```

```python
print(user_details.shape)
print(cooking_sessions.shape)
print(order_details.shape)
```

```python
user_details.info()
```

```python
cooking_sessions.info()
```

```python
order_details.info()
```

```python
order_details.isna().sum()
```

```python
# Handle missing values

order_details.fillna({'Rating': order_details['Rating'].mean()}, inplace=True)
```

## Step 2: Data Merging

```python
# Merge cooking sessions with user details

merged_data = pd.merge(cooking_sessions, user_details, on='User ID', how = 'left')
```

```python
merged_data.head()
```

```python
# Merge with order details

final_data = pd.merge(merged_data, order_details, on=['User ID', 'Session ID'], how = 'left')
```

```python
final_data.head()
```

```python
# Drop duplicate columns
final_data = final_data.drop(columns=['Dish Name_y', 'Meal Type_y'])

# Rename columns to remove _x suffix
final_data.rename(columns={'Dish Name_x': 'Dish Name', 'Meal Type_x': 'Meal Type'}, inplace=True)
```

```python
final_data.head()
```

## Step 3: Export Cleaned Data

```python
# Export the cleaned data
final_data.to_csv('cleaned_data.csv', index=False)
```
